```python
import pandas as pd
# Convert the 'date' column to datetime.date objects using the correct format
sentiment['date'] = pd.to_datetime(sentiment['date'], format='%Y-%m-%d', errors='coerce').dt.date

display(sentiment.head())
```

```python
# Merging data
merged = pd.merge(trades, sentiment[['date', 'classification']], on='date', how='left')

display(merged.head())
```

```python
merged['classification'].isna().sum()

# here we can see the 6 null classifcations.

##Possible reasons:

#Holiday / missing sentiment day
#Timezone edge case
#FearGreed dataset me gap
```

```python
merged_clean = merged.dropna (subset =['classification'])
merged_clean['classification'].isna().sum()
```

##**Dropping Columns**

Dropping columns (`Timestamp IST`, `Transaction Hash`, `Order ID`, `Crossed`, `Trade ID`, and `Timestamp`) as they are either redundant or not directly relevant to our current analytical objectives, which focus on profitability, risk, and trading volume in relation to market sentiment. Removing them will streamline the dataset and improve computational efficiency without impacting the core analysis.

```python
# Dropping other specified columns
columns_to_drop = ['Timestamp IST','Transaction Hash', 'Order ID', 'Crossed', 'Trade ID', 'Timestamp']
merged_clean.drop(columns=columns_to_drop, inplace=True)
merged_clean.head()
```

```python
# Saving cleaned CSV file or summary CSV to the Folder
cleaned_csv_path = '/content/drive/MyDrive/Analytics_Rah/ds_Rahul_Singh/csv_files/merged_clean_data.csv'

merged_clean.to_csv(cleaned_csv_path, index=False)

print(f"Cleaned data saved successfully to: {cleaned_csv_path}")

print("You can find this file in your Google Drive under 'Analytics_Rah/ds_Rahul_Singh/csv_files/'.")
```

## Analyze Profitability and Risk by Sentiment


Calculate the mean and standard deviation of 'Closed PnL' for each sentiment classification to understand both average profitability and the volatility (risk) associated with each sentiment.

To analyze profitability and risk by sentiment, I will group the `merged_clean` DataFrame by 'classification' and calculate the mean and standard deviation of 'Closed PnL' for each group.



```python
sentiment_pnl_stats = merged_clean.groupby('classification')['Closed PnL'].agg(['mean', 'std'])
sentiment_pnl_stats
```

## Analyze Trading Volume by Sentiment


Calculate the total 'Size USD' and 'Size Tokens' for each sentiment classification to understand how trading volume changes with different market sentiments.


To calculate the total 'Size USD' and 'Size Tokens' for each sentiment classification, I will group the `merged_clean` DataFrame by 'classification' and then compute the sum for both 'Size USD' and 'Size Tokens' columns. I will then display the resulting DataFrame.



```python
sentiment_volume_stats = merged_clean.groupby('classification')[['Size USD', 'Size Tokens']].sum()
sentiment_volume_stats
```

## Visualize Sentiment Analysis Results

Create visualizations (e.g., bar charts or grouped bar charts) to display the mean 'Closed PnL', standard deviation of 'Closed PnL', and total trading volume ('Size USD' or 'Size Tokens') for each sentiment classification. Ensure appropriate labels and legends are included for clarity.

To visualize the sentiment analysis results, I will create bar charts for mean 'Closed PnL', standard deviation of 'Closed PnL', total 'Size USD', and total 'Size Tokens' by sentiment classification, as per the instructions.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with multiple subplots
plt.figure(figsize=(18, 12))

# Subplot 1: Mean Closed PnL by Sentiment
plt.subplot(2, 2, 1)
sns.barplot(x=sentiment_pnl_stats.index, y=sentiment_pnl_stats['mean'], palette='viridis', hue=sentiment_pnl_stats.index, legend=False)
plt.title('Mean Closed PnL by Sentiment')
plt.xlabel('Sentiment Classification')
plt.ylabel('Mean Closed PnL')
plt.xticks(rotation=45, ha='right')

# Subplot 2: Standard Deviation of Closed PnL by Sentiment
plt.subplot(2, 2, 2)
sns.barplot(x=sentiment_pnl_stats.index, y=sentiment_pnl_stats['std'], palette='magma', hue=sentiment_pnl_stats.index, legend=False)
plt.title('Standard Deviation of Closed PnL by Sentiment')
plt.xlabel('Sentiment Classification')
plt.ylabel('Standard Deviation of Closed PnL')
plt.xticks(rotation=45, ha='right')

# Subplot 3: Total Trading Volume (USD) by Sentiment
plt.subplot(2, 2, 3)
sns.barplot(x=sentiment_volume_stats.index, y=sentiment_volume_stats['Size USD'], palette='cividis', hue=sentiment_volume_stats.index, legend=False)
plt.title('Total Trading Volume (USD) by Sentiment')
plt.xlabel('Sentiment Classification')
plt.ylabel('Total Size USD')
plt.xticks(rotation=45, ha='right')

# Subplot 4: Total Trading Volume (Tokens) by Sentiment
plt.subplot(2, 2, 4)
sns.barplot(x=sentiment_volume_stats.index, y=sentiment_volume_stats['Size Tokens'], palette='plasma', hue=sentiment_volume_stats.index, legend=False)
plt.title('Total Trading Volume (Tokens) by Sentiment')
plt.xlabel('Sentiment Classification')
plt.ylabel('Total Size Tokens')
plt.xticks(rotation=45, ha='right')

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```
